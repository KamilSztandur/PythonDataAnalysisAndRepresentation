> **Zadanie 4** Narysuj histogramy poszczególnych atrybutów zbioru danych `iris` z podziałem na klasy, oraz macierz wykresów punktowych. Oceń przydatność poszczególnych atrybutów do rozróżniania odmian irysa

```python
filename = "iris.csv"
irisData = pd.read_csv(filename, usecols=range(1,6))

species = set()
for x in irisData['species']:
    species.add(x)

for thisSpecies in species:
    irisDataForThisSpecies = irisData[irisData['species'] == thisSpecies]

    plt.figure(figsize=(10,10), dpi= 60)
    plt.suptitle(thisSpecies)
    
    for i in range(0, len(irisDataForThisSpecies.columns) - 1):
        currentAttribute =  irisDataForThisSpecies.columns[i]
        plt.subplot(2, 2, i + 1)
        
        plt.title(currentAttribute)
        irisDataForThisSpecies[currentAttribute].plot.hist()

sns.pairplot(irisData, kind = "scatter", hue = "species")
plt.show()
```
